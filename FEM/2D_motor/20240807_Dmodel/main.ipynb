```python
""" .Identifierの拡張子がついてるファイルを削除 """
import os

def delete_identifier_files(directory):
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.endswith('.Identifier'):
                os.remove(os.path.join(root, file))

delete_identifier_files('.')
```
